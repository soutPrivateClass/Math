# Simbol atau rumus yang sering digunakan pada machine learning.

- Sigma = Penjumlahan bilangan (Summation)

contoh :

x = ( 2 , 6 , 7 , 3 , 5 )

x1 + x2 + x3 + x4 + x5

x1 + x2 + ..... + xn

```python
# SUMMATION

# MENGGUNAKAN FUNGSI ATAU METHOD BIASA
bilangan1 = [1,2,3,4,5]
sum(bilangan1)

# MENGGUNAKAN FUNGSI ATAU METHOD NUMPY
import numpy as np
np.sum(bilangan1)
```

- Phi = Perkalian bilangan (Products)

contoh :

x = ( 2 , 6 , 7 , 3 , 5 )

x1 * x2 * x3 * x4 * x5

x1 + x2 + ..... * xn



```python
# PRODUCTS

# MENGGUNAKAN FUNGSI ATAU METHOD NUMPY
import numpy as np
bilangan2 = [1,2,3,4,5]
np.prod(bilangan2)
```

- Penjumlahan yang bersifat kumulatif (Cumulative Sum) Menggunakan simbol Sigma

y_1 = x1

y_2 = x1 + x2

y_3 = x1 + x2 + x3

```python
# CUMMULATIVE SUM 

# MENGGUNAKAN FUNGSI ATAU METHOD NUMPY
import numpy as np
from IPython.display import display
import sympy as sym
from matplotlib import pyplot as plt

bilangan3 = [1,2,3,4,5]
hasilSum = np.cumsum(bilangan3)

# ARRAY
bilangan4 = [[1,2,3],[4,5,6]]
hasilArray = np.cumsum(bilangan4) # np.cumsum UNTUK MENGHITUNG NILAI CUMMULATIVE SUMMATION ATAU nilai Sigma

# CONVERT KEDALAM PLOT
plt.plot(bilangan3,'rs-')
plt.plot(hasilSum,'bo-')
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.legend(['B3','Cummulative'], loc="upper left") # LABEL KETERANGAN 
plt.show()
```

- Perkalian (Cumulative Products) Menggunakan simbol Phi

y_1 = x1

y_2 = x1 * x2

y_3 = x1 * x2 * x3

```python
# CUMMULATIVE PRODUCTS

# MENGGUNAKAN FUNGSI ATAU METHOD NUMPY
import numpy as np
from matplotlib import pyplot as plt
import sympy as sym
from IPython.display import display, Math

bilangan5 = [1,2,3,4,5]
hasilCumProd = np.cumprod(bilangan4) # np.prod UNTUK MENGHITUNG NILAI CUMMULATIVE PRODUCT ATAU nilai Phi

plt.plot(bilangan5, "rs-")
plt.plot(hasilCumProd, "bo-")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.legend(["B5","Cummulative Product"], loc="best")
plt.show()
```
