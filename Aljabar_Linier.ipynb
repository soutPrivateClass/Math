# Aljabar Linier.

Konsep persamaan matematika untuk bagaimana cara data berkomunikasi dari data satu dengan yang lainnya :

Beberapa konsep Aljabar Linier diantaranya adalah : 
- Matriks
- Vektor

# Vektor.
Kumpulan angka

Vektor secara aljabar (algebraic vectors) ada 2 jenis :

- Vektor baris (row vectors) = berbentuk horizontal memanjang kesamping
    - [ 2 1 ]
    - [ -21 -3 -54 ]
- Vektor kolom (column vectors) = berbentuk vertikal memanjang kebawah
    - [ 3 4 -5 ]

Implementasi 2 vektor secara geometris ( 2 Dimensi ) :

```python
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings

filterwarnings('ignore')

baris = np.array([2,3])
kolom = np.array([[1],[4]])

plt.plot([0,baris[0]], [0,baris[1]], 'r') # BENTUK PLOT DARI INDEX 0 HINGGA INDEX BARIS INDEX [0] yaitu 2 dan 3
plt.plot([0,kolom[0]], [0,kolom[1]], 'b')
plt.legend(['Row Vectors','Column Vectors'], loc='upper left')
plt.xlabel('Sumbu x')
plt.ylabel('Sumbu y')
plt.axis('square')
plt.axis([-5,5,-5,5]) # Rentangan nilai (xmin,xmax,ymin,ymax)
plt.grid()
plt.show()
```

Kesimpulan :

Secara geometris memiliki hasil yang sama vektor baris dan vektor kolom dimulai dari 0,0, tetapi secara konsep berbeda.

Implementasi 3 vektor secara geometris ( 3 Dimensi ) :

```python
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings

filterwarnings('ignore')

a = np.array([2,3,-3])
b = np.array([-2,2,4])

fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')

ax.plot([0, a[0]], [0, a[1]], [0, a[2]], 'red', linewidth=2)
ax.plot([0, b[0]], [0, b[1]], [0, b[2]], 'blue', linewidth=2)

# GARIS IMAJINER UNTUK KETIGA SUMBU
ax.plot([0,0],[0,0],[-5,5], '--', color=[.6,.6,.6]) # SUMBU IMAJINER X
ax.plot([0,0],[-5,5],[0,0], '--', color=[.6,.6,.6]) # SUMBU IMAJINER Y
ax.plot([-5,5],[0,0],[0,0], '--', color=[.6,.6,.6]) # # SUMBU IMAJINER Z

ax.set_xlim([-5,5])
ax.set_ylim([-5,5])
ax.set_zlim([-5,5])
plt.show()
```

# Menambahkan & mengurangkan vektor.

- Penjumlahan & pengurangan vektor dapat dilakukan jika memiliki dimensi yang sama

Implementasi pada python :

```python
# PENJUMLAHAN VEKTOR :

import numpy as np
import matplotlib.pyplot as plt

vektor1 = np.array([1,3])
vektor2 = np.array([3,1])

# MENGGUNAKAN OPERASI HITUNG BIASA :
hasil1 = vektor1+vektor2

# MENGGUNAKAN FUNGSI NUMPY :
hasil2 = np.add(vektor1,vektor2)

plt.plot([0, vektor1[0]], [0, vektor1[1]], linewidth=2, label='Vektor 1')
plt.plot([0, vektor2[0]], [0, vektor2[1]], linewidth=2, label='Vektor 2')
plt.plot([0, hasil2[0]], [0, hasil2[1]], linewidth=2, label='Return Value')
plt.grid()
plt.legend()
plt.axis('square')
plt.show()
```

Ploting diatas masih menghasilkan plot satu satu dari setiap vektor dan hasil penjumlahan vektor atau masih belum terbentuk hasil transformasi.

Haisl transformasi seperti dibawah ini :

```python
#  HASIL TRANSFORMASI PENJUMLAHAN VEKTOR :

import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import Math

vektor1 = np.array([1,3])
vektor2 = np.array([3,1])

# MENGGUNAKAN OPERASI HITUNG BIASA :
hasil1 = vektor1+vektor2

# MENGGUNAKAN FUNGSI NUMPY :
hasil2 = np.add(vektor1,vektor2)

plt.plot([0, vektor1[0]], [0, vektor1[1]], linewidth=2, label='Vektor 1')
plt.plot([0, vektor2[0]]+vektor1[0], [0, vektor2[1]]+vektor1[1], linewidth=2, label='Vektor 2')
plt.plot([0, hasil2[0]], [0, hasil2[1]], linewidth=2, label='Return Value')
plt.grid()
plt.legend()
plt.axis('square')
plt.show()
```

# Perkalian vektor dengan skalar.

- Skalar adalah simbol yang biasa didefiniskan sebuah integer

Implementasi pada python :

```python
import numpy as np
import matplotlib.pyplot as plt

vektor = np.array([1,2])
skalar = [-0.5,1,1.5,3]

for i in skalar:
    hasil = i*vektor
    plt.plot([0 , hasil[0]], [0, hasil[1]], 'o-', linewidth=2, label='skalar ke- : %g' %i)

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

Kesimpulan :

- Perkalian vektor dengan skalar tidak merubah garis utamanya, namun merubah arah garisnya ( jika minus garis akan berpotensi kebawah & jika positif akan berpotensi keatas )
